As we see there is  no missing data. Now we want to check if there is any outliers.

```python
mean = np.mean(train)
std = np.std(train)

# calculate summary statistics
data_mean, data_std = np.mean(train), np.std(train)
# identify outliers
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off

# identify outliers
outliers = [x for x in train if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))

# remove outliers
outliers_removed = [x for x in train if x > lower and x < upper]
print('Non-outlier observations: %d' % len(outliers_removed))
```

```python
#Feature Engineering
train = train.drop('id', axis=1)
train['hour']=train['hour'].map(lambda x: datetime.strptime(str(x),"%y%m%d%H"))
train['day_of_week']=train['hour'].map(lambda x:  x.weekday())
train['hour']=train['hour'].map(lambda x:  x.hour)
```

**Distribution**

```python
features = ['banner_pos','device_type', 'device_conn_type','hour','day_of_week',"C1",'click']
sns.set_context("notebook",font_scale=0.5)
 
sns.set_style("dark")
 
width = 0.4
 
fig, axes = plt.subplots(nrows=3, ncols=2,figsize=(20,15))
 
for index,feature in enumerate(features[:len(features)-1]):
 
    x = index // 2
 
    y = index % 2   
 
    df3 = train.groupby([feature,'click'], sort=True).size().unstack(fill_value=0)
 
    df3.columns = ['non click','click']
 
    indices = range(len(df3.index.values))
 
    ctr = [float(value)/(df3['non click'].values[i] + value) for i,value in enumerate(df3['click'].values)]
 
    axes[x,y].bar(indices, df3['non click'].values, width, label='non click')
 
    axes[x,y].bar(indices, df3['click'].values, width,bottom=df3['non click'].values,label='click')
 
    axes[x,y].set_xticks(indices)
 
    axes[x,y].set_xticklabels(tuple(df3.index.values))
 
    axes[x,y].set_ylabel('Number of ads / CTRs')
 
    axes[x,y].set_title('CTR for '+feature)
 
    axes[x,y].legend() 
 
    for p in  axes[x,y].patches:
 
        if p.get_height() in df3['non click'].values: 
 
            axes[x,y].annotate(format(100*ctr[np.where(df3['non click'].values == p.get_height())[0][0]], '.1f')+'%', (p.get_x() + p.get_width() / 2., p.get_height()*1.2 ), ha = 'center', va = 'center', 
 
                                   xytext = (0, 10), textcoords = 'offset points')
 
plt.show()

```
